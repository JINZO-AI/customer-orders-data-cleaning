```python
df=df.drop_duplicates()
```

```python
df = df.copy()
#Numeric columns
for col in ['total_amount'] :
    df[col] = df[col].fillna(df[col].median())
#Categorical columns
categorical_cols = ['customer_id','country','category']

df[categorical_cols] = df[categorical_cols].fillna('Unknown')
```

```python
print(df.isnull().sum()[df.isnull().sum() >0 ])
```

```python
for col in ['category','status','country']:
    df[col] = df[col].str.strip().str.lower()

#step 3 : map o canonical values
category_map = {
    'electronics': 'Electronics',
    'clothing': 'Clothing' ,
    'home & garden' : 'Home & Garden',
    'home and garden': 'Home & Garden',
    'sports': 'Sports',
    'books': 'Books',
    'unknown': 'Unknown'
}

status_map = {
    'pending': 'Pending',
    'shipped': 'Shipped',
    'delivered': 'Delivered',
    'cancelled': 'Cancelled'
}

country_map ={
    'united states': 'United States', 'us': 'United States',
    'u.s.a': 'United States', 'usa': 'United States',
    'united kingdom': 'United Kingdom', 'uk': 'United Kingdom','u.k.': 'United Kingdom',
    'australia': 'Australia', 'aus': 'Australia',
    'canada': 'Canada',
    'germany': 'Germany', 'de': 'Germany',
    'france': 'France', 'fr': 'France',
    'italy': 'Italy', 'it': 'Italy',
    'spain': 'Spain', 'es': 'Spain',
    'unknown': 'Unknown'
}

df['category'] = df['category'].map(category_map).fillna(df['category'])
df['status'] = df['status'].map(status_map).fillna(df['status'])
df['country'] = df['country'].map(country_map).fillna(df['country'])
```

```python
df.head(10)
```

```python
df['unit_price']=(
    df['unit_price']
    .astype(str)
    .str.replace(r'[$£,]','',regex=True) #commas catch 1,299.00 style values
    .str.strip()
)
df['unit_price'] = pd.to_numeric(df['unit_price'],errors='coerce')
```

```python
df.head(10)
```

```python
bool_map = {
    'Yes': True, 'yes': True, 'Y': True, 'y': True,
    'TRUE': True, 'True': True, 'true': True, '1': True, 1: True, True: True,
    'No': False, 'no': False, 'N': False, 'n': False,
    'FALSE': False, 'False': False, 'false': False, '0': False, 0: False, False: False
}
df['is_returned'] = df['is_returned'].map(bool_map).fillna(False)
```

```python
df.head(10)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=False,errors='coerce')
# Fix remaining NaT — reload original raw strings for failed rows
df_raw_dates = pd.read_csv("messy_customer_orders.csv", usecols=['order_date'])
df_raw_dates.columns = ['order_date_raw']
df_raw_dates = df_raw_dates.loc[df.index]  # align index after drop_duplicates

mask = df['order_date'].isna()

# Try MM-DD-YYYY format
df.loc[mask, 'order_date'] = pd.to_datetime(
    df_raw_dates.loc[mask, 'order_date_raw'],
    format='%m-%d-%Y',
    errors='coerce'
)

mask2 = df['order_date'].isna()
print(f"NaT before: {mask.sum()}, NaT after: {mask2.sum()}")
# Second pass — fix "21 Jun 2022" and "21-Jun-22" text month formats
mask3 = df['order_date'].isna()

df.loc[mask3, 'order_date'] = pd.to_datetime(
    df_raw_dates.loc[mask3, 'order_date_raw'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)

mask4 = df['order_date'].isna()
print(f"NaT before: {mask3.sum()}, NaT after: {mask4.sum()}")

failed=df['order_date'].isnull().sum()
```

```python
df.head(10)
```

```python
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_day_of_week'] = df['order_date'].dt.day_name()

print(f"NaN in order_year: {df['order_year'].isna().sum()}")
print(f"NaN in order_month: {df['order_month'].isna().sum()}")
print(f"NaN in order_day_of_week: {df['order_day_of_week'].isna().sum()}")
```

```python
# Drop zero quantity rows (no transaction)
df = df[df['quantity'] != 0].copy()

# Recalculate total_amount from scratch
df['total_amount'] = df['quantity'] * df['unit_price']

print(f"Rows after removing zero quantity: {df.shape[0]}")
print(f"Remaining negative quantity (returns): {(df['quantity'] < 0).sum()}")
```
